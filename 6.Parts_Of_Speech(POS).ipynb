```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
#Download resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
#Sample sentence
sentence = "Iron Man Will Return In Avengers Doomsday"
```

```python
#Tokenization
tokens = word_tokenize(sentence)
```

```python
#POS Tagging
tags = pos_tag(tokens)
print("POS Tags: ", tags)
```
