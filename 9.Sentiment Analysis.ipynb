```python
 from textblob import TextBlob
```

```python
# Sample reviews
reviews = [
    "I love this product, it is amazamazing!",
    "This is the worst purchase Iever made.",
    "The product is okay, not great."
]
```

```python
#Sentiment Analysts

for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    if polarity > 8:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(f"Review: {review}\nSentiment: {sentiment} \nPolarity: {polarity}\n")
```
